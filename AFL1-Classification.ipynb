## Data Cleansing

```python
# VMA Data Cleaning
def clean_vma(value):
    try:
        digits = str(value).replace('.', '')
        num_digits = len(digits)

        if num_digits == 3: 
            return float(digits) / 10.0
        elif num_digits >= 4: 
            return float(digits[:4]) / 100.0
        else: 
            return float(digits)
    except (ValueError, TypeError):
        return np.nan 

vma = df.copy()
vma['Visceral Muscle Area (VMA) (Kg)'] = vma['Visceral Muscle Area (VMA) (Kg)'].apply(clean_vma)

cleaned_file_path = "D:/Assignments/Semester-5/ML/AFL1/Data/gallstone_data_cleaned.csv"
vma.to_csv(cleaned_file_path, index=False)

df = vma
```

```python
# GFR Data Cleaning
def clean_gfr(value):
    try:
        stringValue = str(value)
        digits = stringValue.replace('.', '')
        num_digits = len(digits)

        if num_digits >= 6: 
            if digits.startswith('1'):
                return float(digits[:5]) / 100.0
            else:
                return float(digits[:4]) / 100.0
        else: 
            return float(stringValue)
    except (ValueError, TypeError):
        return np.nan
    
gfr = df.copy()
gfr['Glomerular Filtration Rate (GFR)'] = gfr['Glomerular Filtration Rate (GFR)'].apply(clean_gfr)

cleaned_file_path = "D:/Assignments/Semester-5/ML/AFL1/Data/gallstone_data_cleaned.csv"

gfr.to_csv(cleaned_file_path, index=False)

df = gfr
```

```python
# Vitamin D Data Cleaning
def clean_vitamin_d(value):
    try:
        stringValue = str(value)
        digits = stringValue.replace('.', '')
        num_digits = len(digits)

        if num_digits >= 4: 
            return float(digits[:4]) / 100.0
        else: 
            return float(stringValue)
    except (ValueError, TypeError):
        return np.nan 

vd = df.copy()
vd['Vitamin D'] = vd['Vitamin D'].apply(clean_vitamin_d)

cleaned_file_path = "D:/Assignments/Semester-5/ML/AFL1/Data/gallstone_data_cleaned.csv"
vd.to_csv(cleaned_file_path, index=False)

df = vd
```

```python
# Visualizations (VMA, GFR, Vitamin D)
columns_to_plot = ['Visceral Muscle Area (VMA) (Kg)', 'Glomerular Filtration Rate (GFR)', 'Vitamin D']

for col in columns_to_plot:
        plt.figure(figsize=(15, 6))

        # 1. Box Plot
        plt.subplot(1, 2, 1) 
        sns.boxplot(x=df[col])
        plt.title(f'Box Plot of {col}')
        plt.xlabel(col)

        # 2. Histogram
        plt.subplot(1, 2, 2) 
        sns.histplot(df[col], kde=True, bins=30) 
        plt.title(f'Histogram of {col}')
        plt.xlabel(col)

        plt.suptitle(f'Outlier Analysis for {col}', fontsize=16)
        plt.tight_layout(rect=[0, 0.03, 1, 0.95]) 
        plt.show()
```

```python
# Feature Engineering
df['AST_ALT_Ratio'] = df['Aspartat Aminotransferaz (AST)'] / df['Alanin Aminotransferaz (ALT)']

df['TC_HDL_Ratio'] = df['Total Cholesterol (TC)'] / df['High Density Lipoprotein (HDL)']

df['Age_x_Triglyceride'] = df['Age'] * df['Triglyceride']

df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(0, inplace=True) 

print("New features created!")
print(df[['AST_ALT_Ratio', 'TC_HDL_Ratio', 'Age_x_Triglyceride']].head())
```

```python
# Feature Selection
import json

x = df.drop(['patientId', 'Gallstone Status'], axis=1)
y = df['Gallstone Status']

training_columns = x.columns.tolist()

with open('training_columns.json', 'w') as f:
    json.dump(training_columns, f)

print("Training columns saved to training_columns.json")
```

```python
# Data Splitting
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(f"Training set size: {x_train.shape[0]} samples")
print(f"Testing set size: {x_test.shape[0]} samples")
```

```python
# First Test
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

model = RandomForestClassifier(random_state=42)

param_grid = {
    'n_estimators': [50, 100, 200],  
    'max_depth': [None, 10, 20, 30],      
    'min_samples_leaf': [1, 2, 4]       
}

grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=5, n_jobs=-1, verbose=2)
grid_search.fit(x_train_scaled, y_train)

print("Best parameters found: ", grid_search.best_params_)

best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test_scaled)

# Accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"\nNew Model Accuracy: {accuracy * 100:.2f}%")

# Confusion Matrix
conf_matrix = confusion_matrix(y_test, y_pred)
print("\nConfusion Matrix:")
print(conf_matrix)

# Classification Report
class_report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(class_report)
```

```python
# Feature Importance
import matplotlib.pyplot as plt
import seaborn as sns

importances = best_model.feature_importances_
feature_names = x.columns

feature_importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances})

feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)

plt.figure(figsize=(12, 10))
sns.barplot(x='importance', y='feature', data=feature_importance_df.head(15)) # Plot top 15
plt.title('Top 15 Most Important Features for Predicting Gallstones')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
# Testing
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, accuracy_score
from sklearn.utils.class_weight import compute_sample_weight
import joblib

scaler = StandardScaler()
x_train_scaled = scaler.fit_transform(x_train)
x_test_scaled = scaler.transform(x_test)

sample_weights = compute_sample_weight(class_weight='balanced', y=y_train)

best_params = {
    'learning_rate': 0.1,
    'max_depth': 3,
    'n_estimators': 100,
    'subsample': 0.7
}

final_model = XGBClassifier(random_state=42, eval_metric='logloss', **best_params)

final_model.fit(x_train_scaled, y_train, sample_weight=sample_weights)

y_pred = final_model.predict(x_test_scaled)

accuracy = accuracy_score(y_test, y_pred)
print(f"\nFinal Model Accuracy with Balanced Class Weight: {accuracy * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

joblib.dump(final_model, 'gallstone_prediction_model_weighted.joblib')
joblib.dump(scaler, 'scaler_weighted.joblib')
```

```python
# Final Prediction
import joblib
import json

model = joblib.load('gallstone_prediction_model.joblib')
scaler = joblib.load('scaler.joblib')
with open('training_columns.json', 'r') as f:
    training_columns = json.load(f)

test_df = pd.read_csv('D:/Assignments/Semester-5/ML/AFL1/Data/gallstone_test.csv')

print("Preprocessing the new data...")

test_df['Visceral Muscle Area (VMA) (Kg)'] = test_df['Visceral Muscle Area (VMA) (Kg)'].apply(clean_vma)
test_df['Glomerular Filtration Rate (GFR)'] = test_df['Glomerular Filtration Rate (GFR)'].apply(clean_gfr)
test_df['Vitamin D'] = test_df['Vitamin D'].apply(clean_vitamin_d)

test_df['AST_ALT_Ratio'] = test_df['Aspartat Aminotransferaz (AST)'] / test_df['Alanin Aminotransferaz (ALT)']
test_df['TC_HDL_Ratio'] = test_df['Total Cholesterol (TC)'] / test_df['High Density Lipoprotein (HDL)']
test_df['Age_x_Triglyceride'] = test_df['Age'] * test_df['Triglyceride']
test_df.replace([np.inf, -np.inf], np.nan, inplace=True)
test_df.fillna(0, inplace=True)

X_new_test = test_df[training_columns]
X_new_test_scaled = scaler.transform(X_new_test)

predictions = model.predict(X_new_test_scaled)

output_df = pd.DataFrame({
    'patientId': test_df['patientId'],
    'Gallstone Status': predictions
})

output_filename = 'gallstone_predictions.csv'
output_df.to_csv(output_filename, index=False)

print(f"\n✅ Predictions successfully saved to {output_filename}")
print("\nHere's a preview of your saved file:")
print(output_df.head())
```
