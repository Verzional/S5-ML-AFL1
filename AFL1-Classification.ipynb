## Data Cleansing

```python
# VMA Data Cleaning
def clean_vma(value):
    try:
        digits = str(value).replace('.', '')
        num_digits = len(digits)

        if num_digits == 3: 
            return float(digits) / 10.0
        elif num_digits >= 4: 
            return float(digits[:4]) / 100.0
        else: 
            return float(digits)
    except (ValueError, TypeError):
        return np.nan 

vma = df.copy()
vma['Visceral Muscle Area (VMA) (Kg)'] = vma['Visceral Muscle Area (VMA) (Kg)'].apply(clean_vma)

cleaned_file_path = "D:/Assignments/Semester-5/ML/AFL1/Data/gallstone_data_cleaned.csv"
vma.to_csv(cleaned_file_path, index=False)

df = vma
```

```python
# GFR Data Cleaning
def clean_gfr(value):
    try:
        stringValue = str(value)
        digits = stringValue.replace('.', '')
        num_digits = len(digits)

        if num_digits >= 6: 
            if digits.startswith('1'):
                return float(digits[:5]) / 100.0
            else:
                return float(digits[:4]) / 100.0
        else: 
            return float(stringValue)
    except (ValueError, TypeError):
        return np.nan
    
gfr = df.copy()
gfr['Glomerular Filtration Rate (GFR)'] = gfr['Glomerular Filtration Rate (GFR)'].apply(clean_gfr)

cleaned_file_path = "D:/Assignments/Semester-5/ML/AFL1/Data/gallstone_data_cleaned.csv"

gfr.to_csv(cleaned_file_path, index=False)

df = gfr
```

```python
# Vitamin D Data Cleaning
def clean_vitamin_d(value):
    try:
        stringValue = str(value)
        digits = stringValue.replace('.', '')
        num_digits = len(digits)

        if num_digits >= 4: 
            return float(digits[:4]) / 100.0
        else: 
            return float(stringValue)
    except (ValueError, TypeError):
        return np.nan 

vd = df.copy()
vd['Vitamin D'] = vd['Vitamin D'].apply(clean_vitamin_d)

cleaned_file_path = "D:/Assignments/Semester-5/ML/AFL1/Data/gallstone_data_cleaned.csv"
vd.to_csv(cleaned_file_path, index=False)

df = vd
```

```python
# Feature Selection
import json

x = df.drop(['patientId', 'Gallstone Status'], axis=1)
y = df['Gallstone Status']

training_columns = x.columns.tolist()

with open('training_columns.json', 'w') as f:
    json.dump(training_columns, f)

print("Training columns saved to training_columns.json")
```

```python
# Feature Engineering
df['AST_ALT_Ratio'] = df['Aspartat Aminotransferaz (AST)'] / df['Alanin Aminotransferaz (ALT)']

df['TC_HDL_Ratio'] = df['Total Cholesterol (TC)'] / df['High Density Lipoprotein (HDL)']

df['Age_x_Triglyceride'] = df['Age'] * df['Triglyceride']

df.replace([np.inf, -np.inf], np.nan, inplace=True)
df.fillna(0, inplace=True) 

print("New features created!")
print(df[['AST_ALT_Ratio', 'TC_HDL_Ratio', 'Age_x_Triglyceride']].head())
```

```python
# Data Splitting
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)

print(f"Training set size: {x_train.shape[0]} samples")
print(f"Testing set size: {x_test.shape[0]} samples")
```

```python
# Model Tuning, Training, and Saving (Corrected and with original variable names)
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV
from sklearn.utils.class_weight import compute_sample_weight
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score, classification_report
import joblib

pipeline = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', XGBClassifier(random_state=42, eval_metric='logloss', use_label_encoder=False))
])

param_dist = {
    'classifier__n_estimators': [100, 200, 300],
    'classifier__max_depth': [3, 4, 5, 7],
    'classifier__learning_rate': [0.01, 0.05, 0.1, 0.2],
    'classifier__subsample': [0.7, 0.8, 1.0],
    'classifier__colsample_bytree': [0.7, 0.8, 1.0]
}

sample_weights = compute_sample_weight(class_weight='balanced', y=y_train)

random_search = RandomizedSearchCV(
    estimator=pipeline,
    param_distributions=param_dist,
    n_iter=50,
    cv=5,
    n_jobs=-1,
    verbose=2,
    random_state=42
)

random_search.fit(x_train, y_train, classifier__sample_weight=sample_weights)

print("Best parameters found: ", random_search.best_params_)

best_model = random_search.best_estimator_

y_pred = best_model.predict(x_test)

accuracy = accuracy_score(y_test, y_pred)
print(f"\nFinal Tuned Model Accuracy: {accuracy * 100:.2f}%")
print("\nClassification Report:")
print(classification_report(y_test, y_pred))

joblib.dump(best_model, 'model.joblib')
print("\n✅ Complete pipeline model saved to model.joblib")
```

```python
# Feature Importance 
import matplotlib.pyplot as plt
import seaborn as sns

importances = best_model.named_steps['classifier'].feature_importances_
feature_names = x.columns

feature_importance_df = pd.DataFrame({'feature': feature_names, 'importance': importances})
feature_importance_df = feature_importance_df.sort_values(by='importance', ascending=False)

plt.figure(figsize=(12, 10))
sns.barplot(x='importance', y='feature', data=feature_importance_df.head(15)) 
plt.title('Top 15 Most Important Features for Predicting Gallstones')
plt.xlabel('Importance Score')
plt.ylabel('Feature')
plt.tight_layout()
plt.show()
```

```python
# Final Prediction 
import pandas as pd
import numpy as np
import joblib
import json

pipeline = joblib.load('model.joblib')

with open('training_columns.json', 'r') as f:
    training_columns = json.load(f)

test_df = pd.read_csv('D:/Assignments/Semester-5/ML/AFL1/Data/gallstone_test.csv')

test_df['Visceral Muscle Area (VMA) (Kg)'] = test_df['Visceral Muscle Area (VMA) (Kg)'].apply(clean_vma)
test_df['Glomerular Filtration Rate (GFR)'] = test_df['Glomerular Filtration Rate (GFR)'].apply(clean_gfr)
test_df['Vitamin D'] = test_df['Vitamin D'].apply(clean_vitamin_d)

test_df['AST_ALT_Ratio'] = test_df['Aspartat Aminotransferaz (AST)'] / test_df['Alanin Aminotransferaz (ALT)']
test_df['TC_HDL_Ratio'] = test_df['Total Cholesterol (TC)'] / test_df['High Density Lipoprotein (HDL)']
test_df['Age_x_Triglyceride'] = test_df['Age'] * test_df['Triglyceride']
test_df.replace([np.inf, -np.inf], np.nan, inplace=True)
test_df.fillna(0, inplace=True)

X_new_test = test_df[training_columns]

predictions = pipeline.predict(X_new_test)

output_df = pd.DataFrame({
    'patientId': test_df['patientId'],
    'Gallstone Status': predictions
})

output_filename = 'D:/Assignments/Semester-5/ML/AFL1/Data/gallstone_predictions2.csv'
output_df.to_csv(output_filename, index=False)

print(f"Predictions successfully saved to {output_filename}")
print("\nSaved file preview:")
print(output_df.head())
```
